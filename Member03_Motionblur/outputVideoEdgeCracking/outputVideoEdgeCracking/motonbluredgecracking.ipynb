```python
import cv2
import numpy as np
import os


input_folder = "C:/Users/admin/Desktop/outputVideoEdgeCracking"
output_folder = "MotionBlur_Corrected_Frames"


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

print("Output folder ready.")


for filename in os.listdir(input_folder):
    
    if filename.endswith(".jpg"):

        
        image_path = os.path.join(input_folder, filename)
        img = cv2.imread(image_path)

        
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        gray = img

        
        sharpening_kernel = np.array([[0, -1, 0],
                                       [-1, 5, -1],
                                       [0, -1, 0]])

        sharpened_image = cv2.filter2D(gray, -1, sharpening_kernel)

        
        new_filename = "MotionBlurCorrected_" + filename


        save_path = os.path.join(output_folder, new_filename)
        cv2.imwrite(save_path, sharpened_image)

print("Motion blur reduction completed for all frames.")
```
